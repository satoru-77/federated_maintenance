## 2. IMPORTANT REALIZATION

👉 This dataset already contains FULL life of engines

That means:

We know when each engine dies
That’s why we can calculate RUL

```python
rul = df.groupby('engine_id')['cycle'].max().reset_index()
rul.columns = ['engine_id', 'max_cycle']

df = df.merge(rul, on='engine_id')

df['RUL'] = df['max_cycle'] - df['cycle']

print(df[['engine_id', 'cycle', 'max_cycle', 'RUL']].head(10))
```
